#### Mixed integer Linear Programming
Some or all variable are integers
Using Integers in Linear Program because it represents quantities which have to be whole numbers. ie: Not possible to make 3.4 cars, and Integer variables represent decisions so they can only take binary variables

```python
from gurobipy import *
```

Create Model

```python
milp_model = Model('milp')
```

Add Decision Variables with data types

```python
x = milp_model.addVar(vtype=GRB.BINARY, name='x')
y = milp_model.addVar(vtype=GRB.CONTINUOUS, lb=0, name='y')
z = milp_model.addVar(vtype=GRB.INTEGER, lb=0, name='z')
```

Define Objective Function (Maximize)

```python
obj_fn = 2*x + y + 3*z
milp_model.setObjective(obj_fn, GRB.MAXIMIZE)
```

Add the constraints

```python
c1 = milp_model.addConstr(x + 2*y + z <= 4, 'c1')
c2 = milp_model.addConstr(2*z + y <= 5, 'c2')
c3 = milp_model.addConstr(x + y >= 1, 'c3')
```

```python
milp_model.optimize()
```

```python
# Get values for decision variables
for v in milp_model.getVars():
    print('%s: %g' % (v.VarName, v.x))
```
